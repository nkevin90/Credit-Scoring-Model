Unlike numerical data, categorical data doesn't have a natural order or magnitude. Therefore, the concept of outliers, which is often associated with numerical data, doesn't apply directly to categorical data. Hence we check outliers for numerical data only

```python
# Check for Outliers for the original Numeric columns(Does not include categorical columns that have been transformed)
numerical_columns = ['DURATION','AMOUNT','INSTALL_RATE','AGE','NUM_CREDITS','NUM_DEPENDENTS']
numerical_data = data[numerical_columns]
melted_data = pd.melt(numerical_data)

# Create box plots to check outliers
plt.figure(figsize=(12, 8))
sns.boxplot(x="variable", y="value", data=melted_data)
plt.xticks(rotation=45)  
plt.xlabel("Numerical Columns")
plt.ylabel("Count")
plt.title("Box Plot of Original Numerical Columns")
plt.tight_layout()
plt.show()
```

I have identified the presence of outliers in the '**AMOUNT**' column. However, upon closer examination, I believe that these outliers can be explained by other predictor variables. Given the nature of credit score analysis and the potential interactions between different variables, I have decided to exercise caution before categorically treating these outliers as anomalies. The 'amount' column's behavior might be influenced by the dynamics of other variables in the dataset. As a result, I am considering a more comprehensive approach to understanding their impact before making any adjustments or exclusions.

At this point, I will conduct a correlational analysis of the 'AMOUNT' column with the other variables in the dataset to unveil the existence of moderate to strong positive correlations (ranging from 0.6 to 1.0) or negative correlations (ranging from -0.6 to -1.0). This analysis will help me determine if the 'AMOUNT' column's behavior is significantly correlated with any other variable. The presence of such correlations could provide insights into the relationship between the 'amount' column and other predictors, which may help explain the outliers' presence.

By exploring these correlations, I aim to better understand the interplay between variables and evaluate whether the outliers are influenced by other factors. 

```python
# Calculate correlation with 'AMOUNT
correlation_with_amount = data.corr()['AMOUNT'].drop('AMOUNT')

# colormap with a scale of -1 to 1
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Normalize correlation values to match colormap scale
norm = plt.Normalize(correlation_with_amount.min(), correlation_with_amount.max())
correlation_colors = cmap(norm(correlation_with_amount))

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=correlation_with_amount, y=correlation_with_amount.index, palette=correlation_colors)
plt.title('Correlation with "AMOUNT" Bar Plot')
plt.xlabel('Columns')
plt.ylabel('Correlation')
plt.xticks(rotation=45)

# Filter correlations with absolute values between 0.6 and 1.0
strong_correlations = correlation_with_amount[(correlation_with_amount >= 0.6) | (correlation_with_amount <= -0.6)]

# Print strong correlations
for col, val in strong_correlations.items():
    print('Strong Correlations:')
    print(f'{col}: {val:.4f}')

plt.show()
```

# 5. Exploratory Dataset Analysis(EDA)

## a. Univariate Analysis

### What is the distribution of individuals with good credit(Non-Defaulters) and bad credit(Defaulters) in the dataset?

**Assumption**: Given that the dataset lacks records with loan durations exceeding 90 days, I will make the following assumption for the purpose of the analysis:

Applicants with a good credit rating will be categorized as non-defaulters.
Applicants with a bad credit rating will be categorized as defaulters.
This assumption is based on the dataset's maximum recorded loan duration of 72 days, which falls below the 90-day threshold for defining defaulters. As a result, I will align the credit rating with the default status, where a good credit rating signifies non-defaulters and a bad credit rating signifies defaulters.

```python
# Check Distribution of bad credit(0) to good credit(1)
total_records = len(data)
good_credit = data['RESPONSE'].value_counts()[1]
bad_credit = data['RESPONSE'].value_counts()[0]

good_credit_percentage = (good_credit / total_records) * 100
bad_credit_percentage = (bad_credit / total_records) * 100

# Create the pie chart
labels = ['Good Credit (Non-Defaulters)', 'Bad Credit (Defaulters)']
sizes = [good_credit_percentage, bad_credit_percentage]
colors = ['green', 'red']
explode = (0, 0.1)  

plt.figure(figsize=(12, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.2f%%', shadow=True, startangle=90,
        pctdistance=0.85, labeldistance=1.1)
plt.axis('equal')
plt.title('Distribution of Non-Defaulters and Defaulters')
plt.show()
```

The dataset's class distribution reveals a noteworthy class imbalance, with a substantial 70% of instances classified as non-defaulters and a comparatively smaller 30% as defaulters. 

### How are the numerical variables distributed?

```python
# Create KDE plots to show distribution of the dataset
fig, axes = plt.subplots(3, 2, figsize=(12, 8))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Create individual KDE plots for each numerical column
for i, column in enumerate(numerical_data.columns):
    sns.kdeplot(numerical_data[column], shade=True, ax=axes[i])
    
    # Calculate mean and median
    mean_value = np.mean(numerical_data[column])
    median_value = np.median(numerical_data[column])
    
    # Add lines for mean and median
    axes[i].axvline(mean_value, color='r', linestyle='--', label='Mean')
    axes[i].axvline(median_value, color='b', linestyle='--', label='Median')
    
    axes[i].set_title(f'Skewness of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Density')
    axes[i].legend()
    
plt.tight_layout()
plt.show()
```

### How are the Categorical/Binary variables distributed?

```python
categorical_columns = [col for col in data.columns if col not in numerical_columns and col != 'RESPONSE']

# Create bar plots for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(12, 8))
    temp_data = data.copy()
    
    if len(temp_data[column].unique()) == 2:
        temp_data[column] = temp_data[column].replace({0: 'No', 1: 'Yes'})
    
    ax = sns.countplot(temp_data[column], palette='viridis')
    plt.title(f'Bar Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    
    total_count = len(temp_data[column])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total_count)
        ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
    
    plt.tight_layout()
    plt.show()
```
